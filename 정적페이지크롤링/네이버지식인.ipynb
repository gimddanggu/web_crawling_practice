```python
import requests
from bs4 import BeautifulSoup

path = r'https://kin.naver.com/search/list.naver?query=%EC%82%BC%EC%84%B1%EC%A0%84%EC%9E%90'
response = requests.get(path)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 1단계
# 첫번째 글 - 제목, 링크, 날짜, 카테고리, 답변수

a = soup.select_one('dl dt')
print(a.text.strip())   # 제목
print(a.select_one('a')['href'])   # 링크
b = soup.select_one('dl dd')
print(b.text)
# c = soup.select_one('.txt_g1 _nclicks:kin.cat2')
# print(c)
d = soup.select_one('dd a:nth-of-type(2)')
print(d.text)
e = soup.select_one('.hit')
print(e.text.split()[1])
```

### 2단계
- 1페이지 10개 지식인
- 제목, 링크, 날짜, 카테고리, 답변수 크롤링해서 출력

```python
## 나무테그 dl
## .basic1 > dl
elements = soup.select('dl')
for element in elements:
    title = element.select_one('._nclicks\\:kin\\.txt').text
    link = element.select_one('._nclicks\\:kin\\.txt')['href']
    date = element.select_one('.txt_inline').text
    category = element.select_one('.txt_block > a:nth-of-type(2)').text
    answer_num = element.select_one('.txt_block > span:nth-of-type(2)').text.split()[1]
    
    print(title, link, date, category, answer_num, sep='\n')
    print()
```

### 3단계
- 1페이지 ~ N페이지까지 크롤링하기(N: 사용자가 입력한 숫자)
- 힌트) 파이썬 INPUT 명령어 사용해보기

```python
nPage = int(input('몇 페이지까지 출력할까요? > '))

for i in range(1, nPage+1):
    path = fr'https://kin.naver.com/search/list.naver?query=%EC%82%BC%EC%84%B1%EC%A0%84%EC%9E%90&page={i}'
    response = requests.get(path)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    elements = soup.select('dl')
    for element in elements:
        title = element.select_one('._nclicks\\:kin\\.txt').text
        link = element.select_one('._nclicks\\:kin\\.txt')['href']
        date = element.select_one('.txt_inline').text
        category = element.select_one('.txt_block > a:nth-of-type(2)').text
        answer_num = element.select_one('.txt_block > span:nth-of-type(2)').text.split()[1]
        
        print(title, link, date, category, answer_num, sep='\n')
        print()

```

### 4단계
- 크롤링한 데이터 엑셀에 저장하기

```python
import pandas as pd
```

```python
nPage = int(input('몇 페이지까지 출력할까요? > '))

# 데이터를 담을 리스트 생성+
data = []
for i in range(1, nPage+1):
    path = fr'https://kin.naver.com/search/list.naver?query=%EC%82%BC%EC%84%B1%EC%A0%84%EC%9E%90&page={i}'
    response = requests.get(path)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    elements = soup.select('dl')
    for element in elements:
        title = element.select_one('._nclicks\\:kin\\.txt').text
        link = element.select_one('._nclicks\\:kin\\.txt')['href']
        date = element.select_one('.txt_inline').text
        category = element.select_one('.txt_block > a:nth-of-type(2)').text
        answer_num = element.select_one('.txt_block > span:nth-of-type(2)').text.split()[1]
        data.append([title, link, date, category, int(answer_num)]) # 
    
        
```

df 생성
- pd.DataFrame(data, columns=['컬럼1', '컬럼2'..])
df → exel 변환
- DataFrame.to_excel('파일명')

```python
try:
    df = pd.DataFrame(data, columns=['제목', '링크','날짜','카테고리', '답변수'])
    df.to_excel('naver_kin_crawling.xlsx')
    print('성공적으로 파일이 생성되었습니다.')
except Exception as e:
    print(e)
```
