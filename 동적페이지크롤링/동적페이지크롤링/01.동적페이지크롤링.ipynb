```python
# Q. 첫번째 상품의 상품명 추출하기

# 1. 정적 페이지 추출방식
import requests
from bs4 import BeautifulSoup

path = r"https://startcoding.pythonanywhere.com/dynamic"
response = requests.get(path)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
title = soup.select_one(".product-name:nth-of-type(1)")
print(title)
```

```python
# 2. 동적 페이지 추출방식
 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# 크롬 드라이버 생성
driver = webdriver.Chrome()

# 페이지이동
driver.get("https://startcoding.pythonanywhere.com/dynamic")
```

```python
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
soup
```

`<div>` 태그 내용이 채워진 것을 확인할 수 있다.

```python
soup.select_one(".product-name")
driver.find_element(By.CSS_SELECTOR, ".product-name").text
```
